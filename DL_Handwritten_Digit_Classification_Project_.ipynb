#Pre-Process the data

```python
import numpy as np
np.unique(ytrain)
```

```python
#Scaling the train and test data pixel values 0-255
xtrain_scl= xtrain.astype("float32")/255
xtest_scl=xtest.astype("float32")/255
```

```python
from keras.utils import to_categorical
ytrain_cat= to_categorical(ytrain)
ytest_cat= to_categorical(ytest)
```

```python
from keras.models import Sequential
from keras.layers import Input, Conv2D, MaxPool2D, Flatten, Dropout, Dense
```

```python
# Initialize the model
model = Sequential()
# Add input layer
model.add(Input(shape=(28, 28, 1)))
# Add Convolution layer 1
model.add(Conv2D(filters=32, kernel_size=(3,3), activation='relu'))
# Add max pooling layer 1
model.add(MaxPool2D(pool_size=(2,2)))
# Add Convolution layer 2
model.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu'))
# Add max pooling 2
model.add(MaxPool2D(pool_size=(2,2)))
# Add Flatten Layer
model.add(Flatten())
# Add dropout
model.add(Dropout(0.5))
# Add dense layer
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
hist = model.fit(xtrain_scl, ytrain_cat, validation_split=0.1, epochs=20)
```

Plot the learning curve

```python
plt.figure(figsize=(10, 5))
plt.plot(hist.history['loss'], label='Train Loss')
plt.plot(hist.history['val_loss'], label='Validation Loss')
plt.xlabel('Number of Epochs')
plt.ylabel('Categorical Crossentropy')
plt.title('Learning Curve')
plt.legend()
plt.grid()
plt.show()
```

Evalaute the model

```python
model.evaluate(xtrain_scl, ytrain_cat)
```

```python
model.evaluate(xtest_scl, ytest_cat)
```

Predict results for train and test

```python
yprob_train = model.predict(xtrain_scl)
yprob_test = model.predict(xtest_scl)
```

```python
yprob_train[0:5]
```

```python
ypred_train = np.argmax(yprob_train, axis=1)
ypred_test = np.argmax(yprob_test, axis=1)
```

Plot the confusion matrix for test

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cf = confusion_matrix(ytest, ypred_test)
cfd = ConfusionMatrixDisplay(cf)
cfd.plot()
```

Print the classification report

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_test))
```
